## Developing a Model

### Implementation: Define a Performance Metric
What is the performance meteric with which you will determine the performance of your model? Create a function that calculates this performance metric, and then returns the score. 

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    # Your code here:
    return r2_score(y_true, y_predict)
```

### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the traing and test. 

```python
# Your code here
from sklearn.model_selection import train_test_split

X = data.drop(columns = 'medv', axis=1)

# According to the documentation, train_test_split wraps input validation and 
# next(ShuffleSplit().split(X, y)) and application to input data into a single call 
# for splitting (and optionally subsampling) data in a oneliner. Therefore I do not see
# the need for additional shuffling.

X_train, X_test, y_train, y_test = train_test_split(X, data['medv'], test_size=0.2, random_state=1, shuffle=True)
```

----

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
# import sys
# !{sys.executable} -m pip install mlxtend

# http://rasbt.github.io/mlxtend/user_guide/evaluate/bias_variance_decomp/
```

```python
# Five separate RFR here with the given max depths

from sklearn.ensemble import RandomForestRegressor
from mlxtend.evaluate import bias_variance_decomp

score_train_list = []
score_test_list = []
max_dp = [1, 2, 4, 6, 8, 10]

for depth in max_dp:
    # Should I use random_state=1?     
    regr = RandomForestRegressor(max_depth=depth, random_state=1)
    regr.fit(X_train, y_train)
    y_train_pred = regr.predict(X_train)
    y_test_pred = regr.predict(X_test)
    
    print("Max depth: ", depth)
    
    score_train = performance_metric(y_train, y_train_pred)
    score_train_list.append(score_train)
    print("Score_train: ", score_train)
    
    score_test = performance_metric(y_test, y_test_pred)
    score_test_list.append(score_test)
    print("Score_test:", score_test)
    
# The code below gives an error:
#     avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
#         regr, X_train, y_train, X_test, y_test, loss='mse', random_seed=1)

    # The following calculations are from the source code of bias_variance_decomp()    
    main_predictions = np.mean(y_test_pred, axis=0)
    avg_bias = np.sum((main_predictions - y_test)**2) / y_test.size
    avg_var = np.sum((main_predictions - y_test_pred)**2) / y_test_pred.size
    
    print('Average bias: %.3f' % avg_bias)
    print('Average variance: %.3f' % avg_var)
    

# The bias would be the average difference between each model's predictions and the true values.
```

Now, plot the score for each tree on the training set and on the testing set.

```python
# Produce a plot with the score for the testing and training for the different max depths

fig, axs = plt.subplots(ncols=2, figsize=(12,5))

sns.scatterplot(max_dp, score_train_list, ax=axs[0])
axs[0].set(xlabel='max depth', ylabel='r2_score', title='Train set')
axs[0].set(ylim=(0.3, 1))

sns.scatterplot(max_dp, score_test_list, ax=axs[1])
axs[1].set(xlabel='max depth', ylabel='r2_score', title='Test set')
axs[1].set(ylim=(0.3, 1))

plt.show()
```

What do these results tell you about the effect of the depth of the trees on the performance of the model?
