```python
# Import your libraries:

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Check that:
# Data is encoded properly.
# Data falls within the expected range, and use domain knowledge whenever possible to 
# define that expected range.
# Deal with missing data in one way or another: replace it if you can or drop it.
```

```python
iris = pd.read_csv('../iris-data.csv')
iris.head()
```

```python
# Checking stats of numeric columns
iris.describe()

# It seems there are 5 missing values in petal_width_cm column. 
# The mean values seem to be fine, but in order to be sure regarding min-max, I will need
# to plot the data.
```

```python
# Checking dataset shape
iris.shape
```

```python
# Checking data types
iris.dtypes

# Ok, data types are consistent with expectations.
```

```python
# Checking the amount of NaN in the dataset
iris.isna().sum()

# Indeed petal_width_cm column misses 5 values.
```

```python
# Checking values in column 'class'
iris['class'].value_counts()
```

```python
# Class values are not entirely consistent. Iris-versicolor is recorded as versicolor in 5 cases.
# Moreover, Iris-setossa is a typo.

iris.loc[iris['class'] == 'versicolor', 'class'] = 'Iris-versicolor'
iris.loc[iris['class'] == 'Iris-setossa', 'class'] = 'Iris-setosa'
```

```python
# Voila, 'class' values are correct now.
iris['class'].value_counts()
```

```python
# Plotting numeric values
iris.hist(figsize=(7,7))
plt.show()

# It is a bit suspicious that sepal_length_cm has some values very close to zero, 
# while the majority is between 4 and 8cm. Need more plotting.
```

```python
# Plotting numeric values in scatterplot

sns.scatterplot(x="petal_length_cm", y="petal_width_cm", data=iris, hue = "class")
plt.show()

# Petal size clusters based on iris species.
```

```python
# Plotting numeric values in scatterplot

sns.scatterplot(x="sepal_length_cm", y="sepal_width_cm", data=iris, hue = "class")
plt.show()

# Sepal size seems rather consistent among Iris-setosa, but Iris-versicolor has sepal length
# too close to zero when sepals are recorded as relatively wide.
```

```python
# Missing values, all in Iris-setosa species

iris.loc[iris['petal_width_cm'].isnull()]
```

```python
# Calculating mean petal width among Iris-setosa of similar petal length

iris.loc[iris['petal_length_cm'].isin([1.4, 1.5, 1.6])]['petal_width_cm'].mean()
```

```python
# Replacing missing values with the mean petal width, typical for the species

iris.loc[iris['petal_width_cm'].isnull(), 'petal_width_cm'] = 0.24
         
```

```python
# The smallest sepal length values look like mistakes.

versicolor = iris.loc[iris['class'] == 'Iris-versicolor'].sort_values(by=['sepal_length_cm'])
versicolor
```

```python
# Calcualting average sepal length for similar size iris of the same species

versicolor.loc[(versicolor['sepal_width_cm'] > 2.4) & (versicolor['sepal_width_cm'] < 3)]['sepal_length_cm'].mean()
```

```python
# Replacing small values for Iris-versicolor
iris.loc[(iris['sepal_length_cm'] < 1) & (iris['class'] == 'Iris-versicolor') , 'sepal_length_cm'] = 5.43
```

```python
# Sepal size seems much more consistent now.

sns.scatterplot(x="sepal_length_cm", y="sepal_width_cm", data=iris, hue = "class")
plt.show()
```

```python
# Check again the consistency
iris.describe()
```
