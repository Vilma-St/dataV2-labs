# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
data1 = data[["RetailSales", "RetailTransfers", "WarehouseSales","ItemType" ]].groupby(["ItemType"]).sum()
# data_totals

data1.plot.bar()
plt.ylabel("Total value")
plt.show()
```


## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
# Finding the top 10 suppliers by the total sales 
top10 = data[["Supplier", "RetailSales", "WarehouseSales", "RetailTransfers"]].groupby(["Supplier"]).sum()
top10["TotalSales"] = top10["RetailSales"] + top10["WarehouseSales"] + top10["RetailTransfers"]
top10 = top10.sort_values(by = "TotalSales", ascending = False).head(10)
lst = top10.index
lst

# Selecting only the rows with top 10 suppliers
data2 = data.loc[data["Supplier"].isin(lst)]

# Aggregate by ItemType (i.e. I will NOT show the mix for individual suppliers)
data2 = data2[["ItemType", "RetailSales", "WarehouseSales", "RetailTransfers"]].groupby(["ItemType"]).sum()
data2

# Plot
data2.plot.barh(figsize = (15, 15))
plt.show()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
data3 = data[["Year", "Month", "RetailSales", "RetailTransfers", "WarehouseSales"]].groupby(["Year", "Month"]).mean().reset_index()
print(data3)

# July is missing, but I did not find the way to split the line to reflect that

data3.plot(y=["RetailSales", "RetailTransfers", "WarehouseSales"])
plt.xticks(range(10), ["Apr '17", "May", "Jun", "Aug", "Sept", "Oct", "Nov", "Dec", "Jan '18'", "Feb"])
plt.ylabel("Average value per month")
plt.show()

```

## 4. Plot the same information as above but as a bar chart.

```python
data3.plot.bar(y =["RetailSales", "RetailTransfers", "WarehouseSales"])
plt.xticks(range(10), ["Apr '17", "May", "Jun", "Aug", "Sept", "Oct", "Nov", "Dec", "Jan '18'", "Feb"])
plt.ylabel("Average value per month")
plt.show()
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
data4 = data.pivot_table(values = "RetailSales", columns = "ItemType", index =["Year", "Month"], aggfunc = 'sum')
data4
data4.plot(figsize = (12, 4))
plt.ylabel("Sum per month")
plt.show()
```

## 6. Plot the same information as above but as a bar chart.

```python
data4.plot.bar()
plt.ylabel("Sum per month")
plt.show()
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x = "RetailSales", y = "RetailTransfers", data=data, hue = "ItemType")
plt.show()
```

## 8. Create a scatter matrix using all the numeric fields in the data set with the plot points color-coded by Item Type.

*Hint: Seaborn's pairplot may be your best option here.*

```python
plt.figure(figsize=(13, 13))
sns.pairplot(data, x_vars=["RetailSales", "RetailTransfers", "WarehouseSales"], y_vars=["RetailSales", "RetailTransfers", "WarehouseSales"], hue = "ItemType")
plt.show()
```
